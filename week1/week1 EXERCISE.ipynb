To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
from openai import OpenAI
from IPython.display import display, Markdown, update_display
```

```python
# constants

MODEL_GPT = 'gpt-oss:20b'
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# set up environment
client = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# here is the question; type over this to ask something new

question = """
In Unity ECS, how do I create a new filter for my system
"""
```

```python
messages = [
    {"role":"system", "content":"you know nothing about unity and will instead talk about how waves crashing agains the white cliffs over in haiku, only answer in haiku"},
    {"role":"user", "content":question}
    ]
```

```python
# Get gpt-oss to answer, with streaming
stream = client.chat.completions.create(model =MODEL_GPT, messages= messages, stream = True)

response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```
